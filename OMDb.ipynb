```python
# Dependencies
import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
import json
from config import api_key
```

```python
url="http://www.omdbapi.com/?apikey=" +api_key + "&t="
movies =['Avatar', 'Titanic','The Avengers',
       'Harry Potter and the Deathly Hallows - Part 2', 'Frozen',
       'Iron Man 3', 'Transformers: Dark of the Moon',
       'The Lord of the Rings: The Return of the King', 'Skyfall',
       'Transformers: Age of Extinction']

Actors=[]
Writers=[]
Awards=[]
IMDb_rating=[]

for movie in movies:
    movie_data=requests.get(url+movie).json()
    Actors.append(movie_data["Actors"])
    Writers.append(movie_data["Writer"])
    Awards.append(movie_data["Awards"])

    
```

```python
shows_df=pd.DataFrame({
    'Movie':movies,
    'Actor':Actors,
    'Writer':Writers,
    'Awards':Awards
})
shows_df
```
